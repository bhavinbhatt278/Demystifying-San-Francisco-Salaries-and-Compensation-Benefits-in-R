Using 2015 as an example, Fiscal Year 2015 covers the period from July 2014 to June 2015 and Calendar Year 2015 covers the period from January 2015 to December 2015. We will consider Fiscal year for our analysis as hiring, financial and all other statistics for a company are based on fiscal year.

* Department - This column provides us details about various departments that fall under the organization umbrella. We would use this column to predict which department offers higher and lower salaries.

* Job - This column provides us with detailed information about the designation of the employee. This can also provide us with the salary range within a particular job role for a fresher and experienced employee and their growth potential.

* Total.Salary - This column consists of sum of Salary + Overtime + Other benefits from the main dataframe. We weill use this as our primary data to predict various patterns in paygrades among companies.

* Retirement - This column describes city contributions to employee retirement plans. We will use this column data to describe which job has more security as compared to others.

* Total Benefits - This column is  Retirement + Health&Dental Benefits + Other Benefits. It describes the total benefits that a company offers to an employee interms of health, retirement and other benefits.

```r
# Initialize an empty vector to store the results
average_salaries <- numeric()

# Get unique years in the dataset
unique_years <- unique(df2$Year)

# Loop through each unique year
for (year in unique_years) {
  # Subset the dataframe for the current year
  year_data <- df2[df2$Year == year, ]

  # Calculate the mean of "Total.Salary" for the current year
  mean_salary <- mean(year_data$Total.Salary, na.rm = TRUE)

  # Append the result to the vector
  average_salaries <- c(average_salaries, mean_salary)
}

# Create a data frame with year and average salary
result <- data.frame(Year = unique_years, Average_Salary = average_salaries)

# Print the results
print(result)
```

```r
# Create a bar chart
ggplot(result, aes(x = Year, y = Average_Salary)) +
  geom_bar(stat = "identity", fill = "orange", width = 0.5) +
  labs(title = "Average Salary by Year",
       x = "Year",
       y = "Average Salary") +
  theme_minimal()
```

* The plot above describes the average salaries per year for the San Francisco jobs dataset. It can be observed that the average total salaries are more or less same for 2013 to 2015 and then there was a gradual rise in average total salaries after 2016 with 2020 having recorded the highest average salaries.

```r
#Loop 2
C
# Initialize an empty data frame to store the results
lowest_avg_salaries_2020 <- data.frame(Department = character(), Average_Salary = numeric())

# Get unique departments in the dataset
unique_departments <- unique(df2$Department)

# Filter the data for the year 2020
data_2020 <- df2[df2$Year == 2020, ]

# Loop through each unique department
for (department in unique_departments) {
  # Subset the data for the current department
  department_data <- data_2020[data_2020$Department == department, ]

  # Calculate the average salary for the current department
  avg_salary <- mean(department_data$Total.Salary, na.rm = TRUE)

  # Create a new row for the results
  result_row <- data.frame(Department = department, Average_Salary = avg_salary)

  # Append the result row to the lowest_avg_salaries_2020 data frame
  lowest_avg_salaries_2020 <- rbind(lowest_avg_salaries_2020, result_row)
}

# Sort the data frame by average salary in ascending order
lowest_avg_salaries_2020 <- lowest_avg_salaries_2020[order(lowest_avg_salaries_2020$Average_Salary), ]

# Get the top 10 departments with the lowest average salaries
top_10_lowest_avg_salaries <- head(lowest_avg_salaries_2020, 10)

# Print the top 10 departments
print(top_10_lowest_avg_salaries)
```

```r
#Plotting graph

ggplot(top_10_lowest_avg_salaries, aes(Department, Average_Salary, color = Department, fill = Department)) +
  geom_bar(position = "stack", width = 0.6, stat = "identity") +
  coord_flip() +
  geom_text(aes(label = Average_Salary, hjust = -0.03, colour = "black"), size = 3) +

  theme(axis.line = element_line(color = "orange", size = 2)) +
  theme(panel.background = element_blank()) +
  scale_x_discrete() +
  xlab("Department") + ylab("Average_Salary") +
  theme(legend.position = "none", axis.text = element_text(size = 12, face = "bold"),
        plot.title = element_text(size = 18, face = "bold")) +
  ggtitle("Lowest Departments with Average Salaries - Year 2020")

#Increasing the plot size
options(
  repr.plot.width = 12,
  repr.plot.height = 8
)
```

* It can be obseved from the plot above that the departments above provide lowest average salaries as per 2020 data. Thus choosing a career option in any of the above departments would result in lower earning potential.

```r
#Loop 3

# Initialize an empty data frame to store the results
top_10_avg_salaries_by_job_family <- data.frame(Job_Family = character(), Average_Salary = numeric())

# Get unique job families in the dataset
unique_job_families <- unique(df2$Job.Family)

# Loop through each unique job family
for (job_family in unique_job_families) {
  # Subset the data for the current job family
  job_family_data <- df2[df2$Job.Family == job_family, ]

  # Calculate the average salary for the current job family
  avg_salary <- mean(job_family_data$Total.Salary, na.rm = TRUE)

  # Create a new row for the results
  result_row <- data.frame(Job_Family = job_family, Average_Salary = avg_salary)

  # Append the result row to the data frame
  top_10_avg_salaries_by_job_family <- rbind(top_10_avg_salaries_by_job_family, result_row)
}

# Sort the data frame by average salary in descending order
top_10_avg_salaries_by_job_family <- top_10_avg_salaries_by_job_family[order(-top_10_avg_salaries_by_job_family$Average_Salary), ]

# Get the top 10 job families with the highest average salaries
top_10_job_families <- head(top_10_avg_salaries_by_job_family, 10)

# Print the top 10 job families
print(top_10_job_families)
```
